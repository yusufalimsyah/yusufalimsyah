```python
def actions(self, stste):
  """in the leftmost empty column, try all non-conflicting rows."""
  if stste[-1] is not -1:
    return[] 
  else:
    col = state.index(-1)
    return[row for row in range(self.N) if not self.pruning or not self.conflicted(stste, row, col)]
```

```python
def goal_test(self, state):
  """Check if all colums filled, no conflicts"""
  if state[-1] is -1:
    return False
  return not any(self.conflicted(stste, stste[col], col)
                  for col in range(len(state)))
```
